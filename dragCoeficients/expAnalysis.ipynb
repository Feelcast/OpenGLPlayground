```python
import pandas as pd
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt

def linear_function(x, slope, intercept):
    return slope * x + intercept



veldf = pd.read_csv("velocities.csv")
times = veldf["t"]
vx = veldf["vx"]

slope, intercept, r_value, p_value, std_err = linregress(times, vx)




x = np.linspace(0,5000,1000)
y = linear_function(x,slope,intercept)

#physics data
slope = round(slope, 4)
intercept = round (intercept,4)
sigma = 1800/(1280*580)
vprom = np.mean(vx)
mass = 1000
l = 200

coeff = -2000*mass*slope/(sigma*l*vprom*vprom)

print(slope)
print(intercept)
print(r_value)
print(vprom)
print(coeff)
plt.scatter(times,vx, color='orange', label='Simulated velocity')
plt.plot(x,y, color='red', label=f'v = {slope}t + {intercept}')
plt.xlabel('t/[ms]')
plt.ylabel('Vx/[cm/s]')
plt.title('Vx vs t')
plt.legend()
plt.savefig('speed_line.png', dpi=300, bbox_inches='tight')
plt.show()
```
