```python
import numpy as np
import matplotlib.pyplot as plt

def maxwellBoltzmannFunction(energy, mass, speed):
    parameter = np.sqrt(energy/mass)
    normalization = 4*np.pi/((np.sqrt(2*np.pi)*parameter)**3)
    return normalization*speed*speed*np.exp(-speed*speed/(2*parameter*parameter))

maxSpeed = 320
speeds = np.arange(0,maxSpeed,0.1)
distribution = 10000*maxwellBoltzmannFunction(4000,1, speeds)

print(speeds)
print(distribution)
plt.plot(speeds, distribution)
plt.show()
```
